```python
import re
import datetime
import pandas as pd
pd.set_option('display.max_colwidth', -1)
```

```python
# convert to df
data = []

with open('dates.txt', 'r') as file:
    for line in file:
        columns = line.strip().split('\t')
        data.append(columns)

df = pd.DataFrame(data, columns=['id', 'text'])
df.set_index(['id'], inplace=True)
df
```

```python
month_names = {"Jan":"01", "Feb":"02", "Mar":"03", "Apr":"04", "May":"05","Jun":"06", "Jul":"07", "Aug":"08", "Sep":"09", "Oct":"10", "Nov":"11", "Dec":"12"}
```

```python
def normalization(year, month = '01', day = '01'):
    
    # day
    day = '0' + day if len(day) != 2 else day
    
    # month  
    if len(month) == 1:
        month = '0' + month
    elif not month.isdigit():
        month = month_names[month[:3]]
        
    # year
    year = '19' + year if len(year) == 2 else year
    
    normalized_date = '{}-{}-{}'.format(year, month, day)
    
    return normalized_date
```

```python
def find_dates(line):
    # patterns = [r'(0?[1-9]|1[0-2])[\/-](0?[1-9]|[12][0-9]|3[01])[\/-](\d{2}|\d{4})',
    #             r'(\d{1,2})[\/-](\d{1,2})[\/-](\d{4}|\d{2})',
    #             r'(\d{1,2})[\/-](\d{4}|\d{2})',
    #             r'(\d{1,2}) (?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*,? (\d{4}|\d{2})',
    #             r'(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]* (\d{1,2})?,? (\d{4}|\d{2})',
    #             r'(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*,? (\d{4}|\d{2})',
    #             r'(\d{4})'
    #           ]
    
    # these 7 patterns are the best patterns after trying multiple times using "regex 101"
    # need match groups for date, month, year for each pattern
    # here give up using "break" because groups of each pattern are different
    pattern_1 = r'(0?[1-9]|1[0-2])[\/-](0?[1-9]|[12][0-9]|3[01])[\/-](\d{4}|\d{2})'
    pattern_2 = r'(0?[1-9]|1[0-2])[\/-](\d{1,2})[\/-](\d{4}|\d{2})'
    pattern_3 = r'(0?[1-9]|1[0-2])[\/-](\d{4}|\d{2})'
    pattern_4 = r'(\d{1,2}) ((?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*),? (\d{4}|\d{2})'
    pattern_5 = r'((?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*) (\d{1,2})?,? (\d{4}|\d{2})'
    pattern_6 = r'((?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*),? (\d{4}|\d{2})'
    pattern_7 = r'(\d{4})'
    

    if re.search(pattern_1, line['text']):
        match = re.search(pattern_1, line['text'])
        the_date = match.group(0)
        month = match.group(1)
        day = match.group(2)
        year = match.group(3)

        result_1 = normalization(year, month, day)
        return result_1

    elif re.search(pattern_2, line['text']):
        match = re.search(pattern_2, line['text'])
        the_date = match.group(0)
        month = match.group(1)
        day = match.group(2)
        year = match.group(3)

        result_2 = normalization(year, month, day)
        return result_2

    elif re.search(pattern_3, line['text']):
        match = re.search(pattern_3, line['text'])
        the_date = match.group(0)
        month = match.group(1)
        year = match.group(2)

        result_3 = normalization(year, month)
        return result_3

    elif re.search(pattern_4, line['text']):
        match = re.search(pattern_4, line['text'])
        the_date = match.group(0)
        month = match.group(2)
        day = match.group(1)
        year = match.group(3)

        result_4 = normalization(year, month, day)
        return result_4

    elif re.search(pattern_5, line['text']):
        match = re.search(pattern_5, line['text'])
        the_date = match.group(0)
        month = match.group(1)
        day = match.group(2)
        year = match.group(3)

        result_5 = normalization(year, month, day)
        return result_5

    elif re.search(pattern_6, line['text']):
        match = re.search(pattern_6, line['text'])
        the_date = match.group(0)
        month = match.group(1)
        year = match.group(2)

        result_6 = normalization(year, month)
        return result_6

    elif re.search(pattern_7, line['text']):
        match = re.search(pattern_7, line['text'])
        the_date = match.group(0)
        year = match.group(1)

        result_7 = normalization(year)
        return result_7
```

```python
date_format = '%Y-%m-%d'

def push_40d(date):
    # convert yyyy-mm-dd str to datetime.obj
    date = datetime.datetime.strptime(date, date_format)
    # push 40 days
    pushed_date = date + datetime.timedelta(days=40)
    
    # converted back to a string in the format 'yyyy-mm-dd'
    return pushed_date.strftime(date_format) 

df['pushed_date'] = df['normalized_date'].apply(push_40d)
df
```
